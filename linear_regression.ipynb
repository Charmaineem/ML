```python
import numpy as np
import matplotlib.pyplot as plt
```

Two points: a house with 1000sqft sold for 300,000 and a 2000sqft house sold for 500,000

```python
x_train = np.array([1,2])
y_train = np.array([300,500])
print(x_train)
print(y_train)
```

Model function

```python
w = 200
b = 100
print(f"Value of w is {w} and b is {b}")
```

```python
def linear_regression(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
tmp_f_wb = linear_regression(x_train, w,b)
print(tmp_f_wb)
```

```python
plt.plot(x_train,tmp_f_wb, c='b', label='Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title('Housing prices')
plt.ylabel('Prices in 1000s of dollars')
plt.xlabel('Size in 1000 sqft')
plt.show()
```

Prediction

```python
cost_1200sqft = w * 1.2 + b
print(f"cost for a 1200sqft house if {cost_1200sqft}")
```

Linear regression given the following w = np.array([5,0,9]) and b=-4

```python
w2 = np.array([5,0,9])
b2 = -4
```

```python
samples = np.random.rand(1000,3)
```

```python
m = samples.shape[0]
y = np.dot(samples,w2 )+b2

```

```python
extended_samples = np.hstack((samples, np.ones((1000,1))))
print(extended_samples)
```

```python
estimated_w = np.linalg.inv(extended_samples.T.dot(extended_samples)).dot(extended_samples.T).dot(y)
print(estimated_w)
```
